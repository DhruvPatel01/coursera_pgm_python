```python
def are_equal(a, b, precision=1e-4):
    if isinstance(a, (int, float)):
        return (abs(a-b) < precision)
    elif isinstance(a, np.ndarray):
        return (np.allclose(a, b, atol=precision, rtol=0))
    elif isinstance(a, dict):
        common = set(a) & set(b)
        return all([are_equal(a[k], b[k]) for k in common])
    else:
        if len(a) != len(b):
            return False
        return all([are_equal(a[i], b[i]) for i in range(len(a))])
    
```

## Recognize Action

```python
data = loadmat('./data/PA9Data.mat', simplify_cells=True)

for k in ['datasetTrain1', 'datasetTrain2']:
    for actionData in data[k]:
        for action in actionData['actionData']:
            action['marg_ind'] -= 1
            action['pair_ind'] -= 1
            
G = data['G']
G[1:, 1] -= 1

data['labels1'] -= 1

for k in ['datasetTest1', 'datasetTest2']:
    for action in data[k]['actionData']:
        action['marg_ind'] -= 1
        action['pair_ind'] -= 1
    data[k]['labels'] -= 1
```
